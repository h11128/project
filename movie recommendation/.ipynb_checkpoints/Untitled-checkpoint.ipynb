```python
import numpy as np
import random
import functools
from sklearn.cluster import KMeans
from sklearn.neighbors import NearestNeighbors as NN
from scipy.spatial import distance
```

```python
data = np.loadtxt("trainSet")
```

```python
print(data.shape)
print(data[:1])
```

```python
X = data
kmeans = KMeans(init='random', n_clusters=42)
kmeans.fit(X)
labels = kmeans.labels_
```

```python
v = [1 for _ in range(124)]
v = [i/sum(v) for i in v]
my_metric = functools.partial(distance.minkowski, w=v)
```

```python
def findcloset(data,labels):
    similar = []
    neigh = NN(n_neighbors=2,algorithm="auto",  metric=distance.minkowski, n_jobs=-1,p=2)
    for i in range(len(X)):
        point = data[i]
        pointlabel = labels[i]
        clusterpoint = np.where(labels == pointlabel)
        clusterdata = data[clusterpoint]
        neigh.fit(clusterdata)
        try:
            neighbor = neigh.kneighbors([data[i]], n_neighbors=2)
            similar.append(neighbor[1][0][1])
        except:
            similar.append(-1)
    return similar
```

```python
similar1 = findcloset(X,labels)
```

```python
def closet(X):
    similar = []
    neigh = NN(n_neighbors=20,algorithm="auto",  metric=distance.minkowski, n_jobs=-1,p=1)
    neigh.fit(X)
    for i in range(len(X)):
        # return index of closet point
        neighbor = neigh.kneighbors([data[i]], n_neighbors=2)
        similar.append(neighbor[1][0][1])
    return similar
```
