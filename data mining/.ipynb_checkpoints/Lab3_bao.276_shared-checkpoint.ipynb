# Choose an evaluation approach
use Crossvalidation  
because this method can avoid overfitting and give some insight of how the model generalize to unkonwn data, and give a more accurate estimation of prediction performance. It is the most commonly used method for model validation.

```python
X = np.array(data.iloc[:, 1:13]) 
y = np.array(data['class']) 
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, random_state=5)
print(X_train.size,y_test.size)
display(X_train, y_train)
```

# Build and evaluate six classifiers
## knn
the tunable parameter is the number of neighbors

```python
import sklearn.neighbors
"""define: determine hyperpameter, plot the process, according to the method and parameter needs to be changed"""
def process(num,parameter,method):
    warnings.warn("once")
    temp = 0
    all_score = []
    good_parameters = 0
    good_scores = 0
    params = {parameter: num}
    for k, v in params.items():
        for val in v:
            clf = method.set_params(**{k: val})
            method.fit(X_train, y_train)
            scores = cross_val_score(method, X_train, y_train, cv=5, scoring='accuracy')
            all_score.append(scores.mean())
            if scores.mean() > temp:
                temp = scores.mean()
                good_scores = scores
                good_parameters = val
    print("good parameter",parameter,":",good_parameters)
    print("Accuracy: %0.2f (+/- %0.2f)" % (good_scores.mean(), good_scores.std() * 2))
    plt.plot(num, all_score)
    plt.xlabel('parameter change')
    plt.ylabel('mean score of CV')
    plt.show()
knn = sklearn.neighbors.KNeighborsClassifier(n_neighbors=0, 
                                             weights='uniform', algorithm='auto', leaf_size=30, 
                                             p=2, metric='minkowski')
num = list(range(1,100))
process(num,'n_neighbors', knn)
```

```python
"""knn prediction"""
knn = sklearn.neighbors.KNeighborsClassifier(n_neighbors=3, 
                                             weights='uniform', algorithm='auto', leaf_size=30, 
                                             p=2, metric='minkowski')
knn.fit(X_train, y_train)
score = knn.score(X_test, y_test)
print (score)
```

knn generalizes quite well compare to the trainning set performance

##  Desicion Tree

tree.DecisionTreeClassifier

Parameters:
max_depth : int or None, optional (default=None)
The maximum depth of the tree. If None, then nodes are expanded until all leaves are pure or until all leaves contain less than min_samples_split samples.

```python
from sklearn import tree
dtree = tree.DecisionTreeClassifier()
num = list(range(1,100))
process(num,'max_depth',dtree)
```

```python
dtree = tree.DecisionTreeClassifier(max_depth = 62)
dtree.fit(X_train,y_train)
score = dtree.score(X_test,y_test)
print(score)
```

## Naïve Bayes
sklearn.naive_bayes.GaussianNB

Parameters:	
priors : array-like, shape (n_classes,)
Prior probabilities of the classes. If specified the priors are not adjusted according to the data.
var_smoothing : float, optional (default=1e-9)
Portion of the largest variance of all features that is added to variances for calculation stability.

As specified in GaussianNB, priors should be none and tunable parameter is var_smoothing(the default is 1e-09)

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB(priors=None)
num = list(range(1,101))
num = [j*1e-10 for j in num]
process(num,'var_smoothing', gnb)
```

```python
"""So for all the var_smoothing,the reslut is the same and have a higher score"""
gnb = GaussianNB(priors=None, var_smoothing=1e-10)
gnb.fit(X_train, y_train)
score = gnb.score(X_test, y_test)
print (score)
```

Naïve Bayes get accuracy of 0.8809523809523809, it generalize not so bad but not so well

Maximum number of iterations. The solver iterates until convergence (determined by ‘tol’) or this number of iterations. For stochastic solvers (‘sgd’, ‘adam’), note that this determines the number of epochs (how many times each data point will be used), not the number of gradient steps.

# Exceptional work

Experiment with different weights and/or distance metrics for the KNeighborsClassifier. Try at least three different configurations and report your findings.

```python
num = list(range(1,100))

knn1 = sklearn.neighbors.KNeighborsClassifier(n_neighbors=0, 
            weights='uniform', algorithm='auto', leaf_size=30, 
             p=2, metric='minkowski')
knn2 = sklearn.neighbors.KNeighborsClassifier(n_neighbors=0, 
            weights='distance', algorithm='auto', leaf_size=30, 
             p=2, metric='minkowski')
knn3 = sklearn.neighbors.KNeighborsClassifier(n_neighbors=0, 
            weights='uniform', algorithm='auto', leaf_size=30, 
             p=2, metric='manhattan')
knn4 = sklearn.neighbors.KNeighborsClassifier(n_neighbors=0, 
            weights='distance', algorithm='auto', leaf_size=30, 
             p=2, metric='manhattan')


print('Weights=uniform & Metric=Minkowski')
process(num,'n_neighbors', knn1)


print('Weights=distance & Metric=Minkowski')
process(num,'n_neighbors', knn2)


print('Weights=uniform & Metric=manhattan')
process(num,'n_neighbors', knn3)

print('Weights=distance & Metric=manhattan')
process(num,'n_neighbors', knn4)
```

In this experiment, We uesd 'uniform' and 'distance' for parameter 'Weights', and used two ways to measure the metric:'Manhattan' and default 'Minkowski'.

When we compared between 'uniform' and 'distance', 'distance' performed a little bit better(about 0.2 increase). This result means each data in this training set have different weights, and its distribution is chaotic.

As for the metric, 'Manhattan' performed much better than 'Minkowski'(about 0.5 increase).
