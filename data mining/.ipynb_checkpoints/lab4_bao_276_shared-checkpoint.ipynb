```python
"""" clean outliers here by compute Z-score of each value in the column, if abs of Z score bigger than 3 than delete this row"""
data = seeds[(np.abs(stats.zscore(seeds)) < 3).all(axis=1)]
data.info()
```

# 3. Find “natural” clusters
## K-Means


```python
X = np.array(data.iloc[:, 0:8]) 
from sklearn.cluster import KMeans
SSE = []
for K in range(2,12):
    kmeans = KMeans(init='random', n_clusters=K)
    kmeans.fit(X)
    SSE = SSE + [kmeans.inertia_]
    print(kmeans.inertia_)
```

```python
# Plot the SSE values we collected above
fig = plt.figure(figsize=(8, 4), dpi=100)
ax = fig.add_subplot(1, 1, 1)
ax.plot(list(range(2,12)), SSE, marker='.', markersize=10)
ax.set_title('SSE for different values of K')
plt.show()
```

There is a elbow in the plot at K=3, so that appears to be the optimal number of clusters.

```python
kmeans = KMeans(init='random', n_clusters=3)
kmeans.fit(X)
y = kmeans.predict(X)
fig = plt.figure(figsize=(4, 4), dpi=200)
ax = fig.add_subplot(1, 1, 1, aspect=1)
ax.scatter(X[:, 0], X[:, 1], marker='.', s=25, edgecolor='', c=y)
ax.set_title('Original Data Points')
plt.show()
```

## Hierachical
from K-means we can see 3 cluster gives best results, so we use 3 cluster in Hierachical clustering
Now we need to determine linkage

```python
from sklearn.cluster import AgglomerativeClustering
X = np.array(data.iloc[:, 0:8]) 
clustering = AgglomerativeClustering(affinity='euclidean', compute_full_tree='auto',
            connectivity=None, linkage='ward', memory=None, n_clusters=3,
            pooling_func='deprecated').fit(X)

Linkage = ["ward", "complete", "average", "single"]
j = 0
for K in Linkage:
    j = j +1
    clustering.set_params(linkage=K)
    y = clustering.fit_predict(X)
    fig = plt.figure(figsize=(12, 4), dpi=100)
    ax1 = fig.add_subplot(1, 2, 1)
    ax2 = fig.add_subplot(1, 2, 2)
    ax1.set_xlim([-0.1, 1.0])
    vertical_spacing = (j+1)*10 # extra spacing to separate the silhouettes
    ax1.set_ylim([0, len(X) + vertical_spacing])

    silhouette_avg = silhouette_score(X, y)
    print("For K = {}, the average silhouette score is {:.3f}".format(K, silhouette_avg))

    sample_silhouette_values = silhouette_samples(X, y)
    y_lower = 10
    
    for i in range(3):
        ith_cluster_silhouette_values = sample_silhouette_values[y==i]
        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / 3)

        ax1.fill_betweenx(np.arange(y_lower, y_upper),
            0, ith_cluster_silhouette_values,
            facecolor=color, edgecolor=color, alpha=0.7)

        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        y_lower = y_upper + 10  # 10 for the 0 samples

    ax1.set_title("Silhouette Plot")
    ax1.set_xlabel("Silhouette Score")
    ax1.set_ylabel("Cluster Label")

    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([]) 
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    
    colors = cm.nipy_spectral(y.astype(float) / 3)
    ax2.scatter(X[:, 0], X[:, 1], marker='.', s=30, lw=0, alpha=0.7, c=colors, edgecolor='k')



    ax2.set_title("Clustered Data")
    ax2.set_xlabel("Feature space for the 1st feature")
    ax2.set_ylabel("Feature space for the 2nd feature")

    plt.suptitle("Silhouette analysis for hierarchical clustering with linkage = {}".format(K),
                 fontsize=14, fontweight='bold')

    plt.show()

```

As shown above when linkage = average the cluster gives best results

```python

X = np.array(data.iloc[:, 0:8]) 
y = AgglomerativeClustering(affinity='euclidean', compute_full_tree='auto',
            connectivity=None, linkage='average', memory=None, n_clusters=3,
            pooling_func='deprecated').fit_predict(X)

for i in range(1,7):
    fig = plt.figure(figsize=(12, 4), dpi=100)
    ax1 = fig.add_subplot(1, 2, 1)
    ax1.scatter(X[:, 0], X[:, i], marker='.', s=25, edgecolor='', c=y)
    sting = "Feature space for the {}st feature".format(i)
    sting1 = "Feature space for the {}st feature".format(i+1)
    ax1.set_title(sting)

    ax2 = fig.add_subplot(1, 2, 2)
    ax2.scatter(X[:, 0], X[:, i+1], marker='.', s=25, edgecolor='', c=y)
    ax2.set_title(sting1)
    
    

plt.show()
```

## DBSCAN

```python
# Create a NearestNeighbors object, fit the data, then get the 3 nearest neighbors
# The value of K is determined by the SSE for different K plot, choose the elbow value, so K = 3
nbrs = NearestNeighbors(n_neighbors=4).fit(data)
distances, indices = nbrs.kneighbors(data)

# Get the indexes of the array sorted by the 3th nearest neighbor
sorted_ix = distances[:,3].argsort()

# Extract the plot y values
py = distances[sorted_ix,3]
px = list(range(len(distances)))

fig = plt.figure(figsize=(6, 6), dpi=100)
ax = fig.add_subplot(1, 1, 1)
ax.plot(px, py) #, marker='.', s=25, edgecolor='k')
ax.set_title('3th nearest neighbor distance')
ax.set_xlabel('Instances (sorted by Kth nearest neighbor distance)')
ax.set_ylabel('Eps')
ax.grid()
plt.show()
```

```python
X = np.array(data.iloc[:, 0:8]) 
# Create DBSCAN object using eps = 1.06 from the plot above.
# As for the minPts we set a range of number to see which minPts is the best performance.
for i in range(2,9):
    dbest = DBSCAN(eps=1.06, min_samples=i)
    y = dbest.fit_predict(X)
    print("We obtained {} clusters.".format(max(y)+1))
    fig = plt.figure(figsize=(4, 4), dpi=150)
    ax = fig.add_subplot(1, 1, 1, aspect=1)
    ax.scatter(X[:, 0], X[:, 5], marker='.', c=y, s=20, edgecolor='')
    string = "Clustered points with MinPts {}".format(i) 
    ax.set_title(string)
    plt.show()
```

In DBSCAN, we use elbow method to determin the optimal number of the hyperparameter "eps". Then we set another hyperparameter "MinPts" from 2 to 8 and print every plot of cluter. By doing this job, we found that MinPts is not optimal when it is too small, such as 2. And a low minPts means it will build more clusters from noise. From the plots we printed above, when MinPts = 3, we can obtained 3 clusters and noise value is the least.

# 4. Exceptional Work

Use principal component analysis (PCA) or Isometric mapping (Isomap) to reduce the dimensionality of the dataset.

From these 3 methods of clustering, we choose K-Means to do pca and clustering.

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
pca.fit(X)
```

```python
Y = pca.fit_transform(X)
```

```python
SSE = []
for K in range(2,11):
    kmeans = KMeans(init='random', n_clusters=K)
    kmeans.fit(Y)
    SSE = SSE + [kmeans.inertia_]
    print(kmeans.inertia_)
```

```python
# Plot the SSE values we collected above
fig = plt.figure(figsize=(8, 4), dpi=80)
ax = fig.add_subplot(1, 1, 1)
ax.plot(list(range(2,11)), SSE, marker='.', markersize=10)
ax.set_title('SSE for different values of K')
plt.show()
```

In the plot, we could see that 3 is the elbow value, so we let K = 3

```python
kmeans = KMeans(init='random', n_clusters=3)
kmeans.fit(Y)
y = kmeans.predict(Y)
fig = plt.figure(figsize=(4, 4), dpi=150)
ax = fig.add_subplot(1, 1, 1, aspect=1)
ax.scatter(Y[:, 0], Y[:, 1], marker='.', s=25, edgecolor='', c=y)
ax.set_title('PCA Data Points')
plt.show()
```
