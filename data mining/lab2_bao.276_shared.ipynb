## Load and visualize the data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.model_selection
from sklearn.neighbors import KNeighborsClassifier
```

```python
lab = pd.read_csv('lab2b.csv')
x = list(lab.a1)
y = list(lab.a2)
classes = list(lab.c)
unique = list(set(classes))
colors = [plt.cm.jet(float(i*0.9)) for i in unique]

for i, u in enumerate(unique):
    xi = [x[j] for j  in range(len(x)) if classes[j] == u]
    yi = [y[j] for j  in range(len(y)) if classes[j] == u]
    plt.scatter(xi, yi, c=colors[i], label=str(u))
plt.xlabel('a1')
plt.ylabel('a2')
plt.legend(loc=2)
plt.show()
```

## Prepare the data

```python
X = np.array(lab.iloc[:, 0:2]) 	# end index is exclusive
y = np.array(lab['c']) 	# another way of indexing a pandas df
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, random_state=42)
display(X_train, y_train)
```

## Train a classifier and classify the test records

```python
knn = KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train, y_train)
y_hat = knn.predict(X_test)
score = knn.score(X_test, y_test)
print (score)
```

```python
score_train = knn.score(X_train, y_train)
print (score_train)
```

```python
plt.figure(figsize=(10,6), dpi=100)
i=0
#print predict value points:
for x in X_test: 
    if y_hat[i] == 0:
        plt.scatter(X_test[i,0], X_test[i,1], marker='.',color='blue', alpha='0.6', label='0')
    if y_hat[i] == 1:
        plt.scatter(X_test[i,0], X_test[i,1], marker='.',color='red', alpha='0.6', label='1')
    i=i+1
        
i=0
#print true value points:
for x in X_test:
    if y_test[i] == 0:
        plt.scatter(X_test[i,0], X_test[i,1], marker='.',color='blue', alpha='0.6', label='0')
    if y_test[i] == 1:
        plt.scatter(X_test[i,0], X_test[i,1], marker='.',color='red', alpha='0.6', label='1')
    i=i+1
    
plt.xlabel('a1')
plt.ylabel('a2')
plt.title("True Values and Predict Values")
plt.show()
```

```python
#calculate the accuracy of prediction:
i=0
correct=0
total=0
for x in y_test:
    total=total+1
    if (y_hat[i] == y_test[i]):
        correct=correct+1
    i=i+1
        

print (correct/total)
```

## Conclusion

The accuracy score from KNN model is 0.856. The accuracy score from the training set is 0.832. KNN prediction is better than training set prediction. There may be several reasons for this, first of all, our data set quantity is only 500, and 375 for training set, which is not a large number. Secondly, for the training set data, there may be some noise data which make the prediction not very good, in contrast, knn model is not very sensitive to noise values.
